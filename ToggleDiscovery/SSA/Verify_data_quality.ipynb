```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ary = np.load('toggleAry.npy')
```

```python
diff = (ary[:,:,0]-ary[:,:,1])
diff2 = np.array([(t[0]-t[1])/(t[0]+t[1]+1e-8) for traj in ary for t in traj]).reshape(10000,1250)
```

### Raw differences in  repressor counts (first 100 trajectories)

```python
fig, ax = plt.subplots(10,10,figsize=(18,18),sharex=True,sharey=True)
[ax[int(np.floor(i/10)),int(np.mod(i,10))].plot(np.arange(0,250,0.2),d) for i, d in enumerate(diff[1:101])];
[ax[int(np.floor(i/10)),int(np.mod(i,10))].plot(np.arange(0,250,0.2),np.zeros(1250)) for i in range(100)];
```

### Normalized differences in repressor abundance (first 100 trajectories)

```python
fig, ax = plt.subplots(10,10,figsize=(18,18),sharex=True,sharey=True)
[ax[int(np.floor(i/10)),int(np.mod(i,10))].plot(np.arange(0,250,0.2),d) for i, d in enumerate(diff2[1:101])];
```

### Distributions of normalized differences for each hour of simulated time across 10k trajectories

```python
fig, ax = plt.subplots(25,10,figsize=(18,48),sharex=True)
for r in range(25):
    for c in range(10):
        sns.distplot(diff2[:,5*(10*r+c)],ax=ax[r,c],kde=False);
```

### Distributions of each repressor count for each hour of simulated time

```python
fig, ax = plt.subplots(25,10,figsize=(18,48),sharex=True)
for r in range(25):
    for c in range(10):
        sns.distplot(ary[:,5*(10*r+c),0],ax=ax[r,c],hist=False);
        sns.distplot(ary[:,5*(10*r+c),1],ax=ax[r,c],hist=False);
```
